---

### üîπ You probably ran something like this:

```python
class MyClass:
    def reversearray(self, arr):
```

---

### üß† Why does `None` appear?

Let‚Äôs break it down:

1. Inside your method:

   ```python
   def reversearray(self, arr):
       print(arr[::-1])
   ```

   * The `print()` statement **displays** the reversed array on the screen.
   * But it **does not return anything**.

2. In Python, every function that doesn‚Äôt explicitly `return` something will **implicitly return `None`**.

   So when you do this:

   ```python
   print(obj.reversearray([5, 6, 2, 3, 4, 1]))
   ```

   The steps are:

   * `reversearray()` prints `[1, 4, 3, 2, 6, 5]`
   * Then, since it doesn‚Äôt return anything, Python automatically returns `None`
   * Then your outer `print()` prints that returned value (`None`)

---

### ‚úÖ How to fix it

If you only want to **see the reversed list**, call it like this:

```python
obj.reversearray([5, 6, 2, 3, 4, 1])
```

Output:

```
[1, 4, 3, 2, 6, 5]
```

(No `None`, because we didn‚Äôt wrap it inside another `print()`)

---

If you want the **function to return** the reversed array so you can use it later, modify it:

```python
class MyClass:
    def reversearray(self, arr):
        return arr[::-1]  # return instead of print
```

Then:

```python
obj = MyClass()
result = obj.reversearray([5, 6, 2, 3, 4, 1])
print(result)
```

Output:

```
[1, 4, 3, 2, 6, 5]
```

No `None`, because now the function returns the reversed list, which you explicitly print.

---

### ‚úÖ Summary

| Case                  | Function Code                          | Output                      |
| --------------------- | -------------------------------------- | --------------------------- |
| Uses `print()` inside | Prints reversed array + returns `None` | `[1,4,3,2,6,5]` then `None` |
| Uses `return` instead | Returns reversed array                 | `[1,4,3,2,6,5]` only        |

Let‚Äôs go step-by-step üëá

---

üö´ Problem with `arr[::-1]`

```python
arr = [5, 6, 2, 3, 4, 1]
arr[::-1]
print(arr)
```

Output:

```
[5, 6, 2, 3, 4, 1]  # unchanged
```

Because `arr[::-1]` returns a *new* reversed list, but doesn‚Äôt modify `arr`.

---

## ‚úÖ Correct In-Place Reversal Approach

You can reverse the list **in place** using **two-pointer swapping**:

### üîπ Solution:

```python
def reversearray(self, arr):
    left = 0
    right = len(arr) - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]  # swap
        left += 1
        right -= 1
```

### üîπ Example Run:

```python
arr = [5, 6, 2, 3, 4, 1]
obj = MyClass()
obj.reversearray(arr)
print(arr)
```

Output:

```
[1, 4, 3, 2, 6, 5]
```

‚úÖ Now the **original array is modified in place**.

---

## üß† What‚Äôs Happening Internally

| Step                | Array State        | Left | Right    |
| ------------------- | ------------------ | ---- | -------- |
| Start               | [5, 6, 2, 3, 4, 1] | 0    | 5        |
| Swap arr[0], arr[5] | [1, 6, 2, 3, 4, 5] | 1    | 4        |
| Swap arr[1], arr[4] | [1, 4, 2, 3, 6, 5] | 2    | 3        |
| Swap arr[2], arr[3] | [1, 4, 3, 2, 6, 5] | 3    | 2 (stop) |

---

## üß© Alternative (Built-in Method)

Python has a **built-in in-place reverse**:

```python
arr.reverse()
```

This modifies the original array directly, just like our loop.

Example:

```python
arr = [5, 6, 2, 3, 4, 1]
arr.reverse()
print(arr)
# Output: [1, 4, 3, 2, 6, 5]
```

##Alternates in an Array

You are given an array arr[], the task is to return a list elements of arr in alternate order (starting from index 0).

Examples:

Input: arr[] = [1, 2, 3, 4]
Output: 1 3
Explanation:
Take first element: 1
Skip second element: 2
Take third element: 3
Skip fourth element: 4

###SOLUTIONN
Perfect üëç ‚Äî this is a simple and common array manipulation question.
Let‚Äôs go step-by-step through the logic and solution.

---

## üß© Problem

You are given an array:

```python
arr = [1, 2, 3, 4]
```

You need to **take every alternate element**, starting from index 0.

That means:

* Take arr[0] ‚Üí 1
* Skip arr[1] ‚Üí 2
* Take arr[2] ‚Üí 3
* Skip arr[3] ‚Üí 4

‚úÖ **Output:** `[1, 3]`

---

## üîπ Approach 1: Using Python slicing

You can use **step slicing** `[::2]` to pick every 2nd element starting from index 0.

### ‚úÖ Code:

```python
def alternate_elements(arr):
    return arr[::2]
```

### ‚úÖ Example:

```python
arr = [1, 2, 3, 4]
print(alternate_elements(arr))  # Output: [1, 3]
```

---

## üîπ Approach 2: Using a simple loop

If you want to do it manually (e.g., for an interview where slicing may not be allowed):

### ‚úÖ Code:

```python
def alternate_elements(arr):
    result = []
    for i in range(0, len(arr), 2):  # step of 2
        result.append(arr[i])
    return result
```

### ‚úÖ Example:

```python
arr = [1, 2, 3, 4]
print(alternate_elements(arr))  # Output: [1, 3]
```

---

## üîπ Approach 3: Using List Comprehension

Pythonic and concise:

```python
def alternate_elements(arr):
    return [arr[i] for i in range(0, len(arr), 2)]
```

---

## ‚öôÔ∏è Time & Space Complexity

| Approach             | Time Complexity | Space Complexity | Notes                        |
| -------------------- | --------------- | ---------------- | ---------------------------- |
| Slicing (`arr[::2]`) | O(n)            | O(n/2)           | Easiest and cleanest         |
| For-loop             | O(n/2)          | O(n/2)           | Works without Python slicing |
| List comprehension   | O(n/2)          | O(n/2)           | Pythonic and readable        |

---

## ‚úÖ Final Example

```python
arr = [10, 20, 30, 40, 50, 60]
print(alternate_elements(arr))
```

**Output:**


##Array search

Given an array, arr[] of n integers, and an integer element x, find whether element x is present in the array. Return the index of the first occurrence of x in the array, or -1 if it doesn't exist.

Examples:

Input: arr[] = [1, 2, 3, 4], x = 3
Output: 2
Explanation: For array [1, 2, 3, 4], the element to be searched is 3. Since 3 is present at index 2, the output is 2.
Input: arr[] = [10, 8, 30, 4, 5], x = 5
Output: 4
Explanation: For array [10, 8, 30, 4, 5], the element to be searched is 5 and it is at index 4. So, the output is 4.
Input: arr[] = [10, 8, 30], x = 6
Output: -1
Explanation: The element to be searched is 6 and it is not present, so we return -1.
Constraints:
1 ‚â§ arr.size ‚â§ 106
0 ‚â§ arr[i] ‚â§ 106
0 ‚â§ x ‚â§ 105

###solution

Excellent üëè ‚Äî this is a classic **linear search** problem.
Let‚Äôs go step-by-step through the explanation and all possible solutions.

---

## üß© Problem Summary

Given:

* An array `arr[]` of integers.
* A value `x` to search for.

You must:

* Find whether `x` exists in the array.
* If it exists ‚Üí return **its first index**.
* If not ‚Üí return **-1**.

---

## üß† Example

### Input:

```python
arr = [1, 2, 3, 4]
x = 3
```

### Output:

```
2
```

Because `3` is found at index `2` (0-based index).

---

## ‚úÖ Approach 1: Linear Search (Simple Loop)

We just traverse the array one element at a time until we find `x`.

### Code:

```python
def find_element(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i  # Return the first index where x is found
    return -1  # If x not found
```

### Example:

```python
arr = [10, 8, 30, 4, 5]
x = 5
print(find_element(arr, x))  # Output: 4
```

---

## ‚úÖ Approach 2: Using Python‚Äôs `in` and `index()` methods

Python has built-in methods that make this easier.

### Code:

```python
def find_element(arr, x):
    if x in arr:
        return arr.index(x)  # returns first occurrence
    else:
        return -1
```

### Example:

```python
arr = [1, 2, 3, 4]
x = 3
print(find_element(arr, x))  # Output: 2
```

---

## ‚úÖ Approach 3: Using List Comprehension (Less common)

If you want to find *all* occurrences but only return the first, you can do:

```python
def find_element(arr, x):
    indices = [i for i, val in enumerate(arr) if val == x]
    return indices[0] if indices else -1
```

---

## üïí Time & Space Complexity

| Approach             | Time Complexity | Space Complexity | Notes                    |
| -------------------- | --------------- | ---------------- | ------------------------ |
| Linear search (loop) | O(n)            | O(1)             | Best for unsorted arrays |
| `in` + `index()`     | O(n)            | O(1)             | Pythonic and clean       |
| List comprehension   | O(n)            | O(n)             | For multiple occurrences |

---

## ‚öôÔ∏è Edge Cases

| Input                       | Expected Output | Explanation           |
| --------------------------- | --------------- | --------------------- |
| `arr = [10, 8, 30], x = 6`  | `-1`            | Not found             |
| `arr = [5, 5, 5, 5], x = 5` | `0`             | First occurrence only |
| `arr = [], x = 10`          | `-1`            | Empty array           |
| `arr = [0, 1, 2, 3], x = 0` | `0`             | Found at first index  |

---

## üß© Final Example

```python

x = 5
```

```python

```

‚úÖ **Time Complexity:** O(n)
‚úÖ **Space Complexity:** O(1)

### Solution

Excellent üëè ‚Äî this is another classic array check question:
**Determine whether the array is sorted in non-decreasing order.**

Let‚Äôs go step-by-step through it üëá

---

## üß© Problem Definition

You are given an array `arr[]`.

üëâ Your task is to check if it is **sorted in non-decreasing order**,
i.e., each element should be **less than or equal to** the next one.

Formally:

```
arr[i] <= arr[i+1]   for all i in range(len(arr) - 1)
```

If this condition holds true for all elements ‚Üí return `True`
Otherwise ‚Üí return `False`.

---

## ‚úÖ Example 1

```
Input: arr = [10, 20, 30, 40, 50]
Output: True
```

‚úÖ Because each element ‚â§ next one.

---

## ‚úÖ Example 2

```
Input: arr = [90, 80, 100, 70, 40, 30]
Output: False
```

‚ùå Because 90 > 80 ‚Üí violates the non-decreasing rule.

---

## ‚úÖ Approach 1: Simple Loop (Best for Interviews)

### Code:

```python
def is_sorted(arr):
    for i in range(len(arr) - 1):
        if arr[i] > arr[i + 1]:
            return False
    return True
```

### Example:

```python
arr = [10, 20, 30, 40, 50]
print(is_sorted(arr))  # Output: True
```

‚úÖ **Time Complexity:** O(n)
‚úÖ **Space Complexity:** O(1)

---

## ‚úÖ Approach 2: Using `all()` and Generator Expression (Pythonic Way)

This is a concise and efficient Python one-liner.

### Code:

```python
def is_sorted(arr):
    return all(arr[i] <= arr[i + 1] for i in range(len(arr) - 1))
```

### Example:

```python
arr = [90, 80, 100, 70, 40, 30]
print(is_sorted(arr))  # Output: False
```

‚úÖ **Same time complexity:** O(n)
‚úÖ **More readable for experienced Python users**

---

## ‚úÖ Approach 3: Compare with Sorted Version (Not optimal for large arrays)

### Code:

```python
def is_sorted(arr):
    return arr == sorted(arr)
```

‚ö†Ô∏è Sorting takes **O(n log n)**, so it‚Äôs slower than O(n) loop-based checks.
Use this only for small datasets or quick scripts.

---

## üß† Edge Cases

| Input                | Expected Output | Explanation                             |
| -------------------- | --------------- | --------------------------------------- |
| `[10]`               | `True`          | Single element is trivially sorted      |
| `[1, 1, 1, 1]`       | `True`          | Equal elements allowed (non-decreasing) |
| `[5, 4, 3, 2, 1]`    | `False`         | Strictly decreasing                     |
| `[-5, -2, -2, 0, 3]` | `True`          | Works with negatives                    |

---

## ‚úÖ Final Example

